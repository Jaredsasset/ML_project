### 추천 시스템

```python
# 함수화
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

def recommended(user_height, user_weight, user_gender, user_tag, top_n, want_size, df):
    
    # 남자
    if user_gender in ['남성', '남', 'male', '남자']:
        use_df = df[df.제품성별 == '남']
        if want_size != 'X':    # 사이즈 선택
            if want_size == 'XS':
                use_df = use_df[use_df.pred_size == 'S']
            else:
                use_df = use_df[use_df.pred_size == want_size]
    # 여자
    else:
        use_df = df[df.제품성별 == '여']
        if want_size != 'X':     # 사이즈 선택
            if want_size in ['XL', '2XL', '3XL']: # 큰 사이즈 선택 - 전체 데이터
                use_df = df.copy()
            else:
                use_df = use_df[use_df.pred_size == want_size]
        
            
        

    ## 신체정보 유사도 측정--------

    # 유저 특성 벡터 생성
    user_vector = [[user_height, user_weight, 1 if user_gender == '여성' else 0]]

    # 특성 벡터 생성
    features = use_df[['키', '몸무게', '유저성별']].replace({'남성': 0, '여성': 1}).values

    # 코사인 유사도 계산
    cosine_sim = cosine_similarity(user_vector, features)

    # 유사도 기준으로 추천 상품 선택
    top_indices = cosine_sim.argsort().flatten()[::-1]
    recommended_products = use_df.iloc[top_indices]

    
    ## 태그 유사도 측정------------

    # 사이즈 기준으로 추출된 데이터프레임 (상위 20개)
    item_df = recommended_products.head(20)

    # 필요한 특성 벡터화
    tfidf = TfidfVectorizer(analyzer = 'char', ngram_range=(1, 2))
    item_features = item_df.태그.replace('#', '')
    item_matrix = tfidf.fit_transform(item_features).toarray()

    # 사용자 특성 벡터 생성
    user_vector = tfidf.transform([user_tag])

    # 코사인 유사도 계산
    similarities = cosine_similarity(user_vector, item_matrix).flatten()

    # 유사도가 가장 높은 인덱스 선택
    top_indices = similarities.argsort()[::-1]  

    # 최종 5개 출력
    item_df = item_df.iloc[top_indices].drop_duplicates('제품명').head(top_n)
    return item_df    
```

```python
print('┌--------------------------------------------------------------------------------┐')
print('    카테고리 → 반팔, 기본, 무지, 로고, 그래픽, 셔츠, 오버핏, 니트, 카라, 루즈핏  ')
print('    사이즈   → XS(여성전용사이즈), S, M, L, XL, 2XL, 3XL  ')
print('└--------------------------------------------------------------------------------┘')
tag = input('추천받을 카테고리를 입력하세요.\t\t\t: ')
heigth = int(input('고객님의 키(cm)를 입력하세요.\t\t\t: '))
weight = int(input('고객님의 몸무게(kg)를 입력하세요.\t\t: '))
gender = input('고객님의 성별(남성, 여성)을 입력하세요.\t\t: ')
want_size = input('고객님의 평소 사이즈를 입력하세요.(없으면 X)\t: ')
item_cnt = int(input('추천받을 상품 개수를 입력하세요.\t\t: '))
# heigth, weight, gender, tag, item_cnt = 170, 65, '남성', '반팔', 10

print(f'\n◁ 카테고리 ▷ [ {tag} ]')
print('====================================================================================================')
print(f'키 : {heigth}cm, 몸무게 : {weight}kg, 성별 : {gender}')
print('고객님께 추천드리는 제품입니다.')
print('====================================================================================================\n')

# df : 사용할 데이터
result = recommended(heigth, weight, gender, tag, item_cnt, want_size, df_)
result.reset_index(drop=True, inplace=True)

for idx, product in result.iterrows():
    print(f"< {idx+1}. {product['제품명']} >\n")
    print(f"브랜드\t: {product['브랜드']}")
    print(f"제품성별\t: {product['제품성별']}")
    if want_size !='X':
        print(f"선택사이즈\t: {want_size}")
    print(f"추천사이즈\t: {product['사이즈']}")
    print(f"카테고리\t: {product['태그']}")
    print(f"링크\t: {product['링크']}\n")
    print('-'*100)
    print()
```
